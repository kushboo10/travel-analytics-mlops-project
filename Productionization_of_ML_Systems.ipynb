## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# -------------------------------
# DATA WRANGLING / PREPROCESSING
# -------------------------------

# 1. Remove duplicate rows
users_df = users_df.drop_duplicates()
flights_df = flights_df.drop_duplicates()
hotels_df = hotels_df.drop_duplicates()

# 2. Handle missing values
# Users dataset: fill missing 'age' with median, 'gender' with mode
users_df['age'].fillna(users_df['age'].median(), inplace=True)
users_df['gender'].fillna(users_df['gender'].mode()[0], inplace=True)
users_df['company'].fillna('Unknown', inplace=True)

# Flights dataset: fill missing numeric values with median
flights_df['price'].fillna(flights_df['price'].median(), inplace=True)
flights_df['distance'].fillna(flights_df['distance'].median(), inplace=True)
flights_df['time'].fillna(flights_df['time'].median(), inplace=True)

# Flights dataset: fill missing categorical values with 'Unknown'
flights_df['from'].fillna('Unknown', inplace=True)
flights_df['to'].fillna('Unknown', inplace=True)
flights_df['flightType'].fillna('Economy', inplace=True)
flights_df['agency'].fillna('Unknown', inplace=True)

# Hotels dataset: fill missing numeric values with median
hotels_df['price'].fillna(hotels_df['price'].median(), inplace=True)
hotels_df['total'].fillna(hotels_df['total'].median(), inplace=True)
hotels_df['days'].fillna(hotels_df['days'].median(), inplace=True)

# Hotels dataset: fill missing categorical values with 'Unknown'
hotels_df['name'].fillna('Unknown', inplace=True)
hotels_df['place'].fillna('Unknown', inplace=True)

# 3. Convert date columns to datetime type
flights_df['date'] = pd.to_datetime(flights_df['date'], errors='coerce')
hotels_df['date'] = pd.to_datetime(hotels_df['date'], errors='coerce')

# 4. Reset index after dropping duplicates
users_df.reset_index(drop=True, inplace=True)
flights_df.reset_index(drop=True, inplace=True)
hotels_df.reset_index(drop=True, inplace=True)

# 5. Optional: check for remaining missing values
print("Users Dataset Missing Values:\n", users_df.isnull().sum())
print("\nFlights Dataset Missing Values:\n", flights_df.isnull().sum())
print("\nHotels Dataset Missing Values:\n", hotels_df.isnull().sum())
````

### What all manipulations have you done and insights you found?

After performing data wrangling and preprocessing, here’s what has been done and the key insights:

1. Duplicates Removal

Manipulation: Removed duplicate rows from users_df, flights_df, and hotels_df.

Insight: There were a few repeated records which could have biased analysis and model predictions. Removing duplicates ensures clean and reliable data.

2. Handling Missing Values

Manipulation:

Filled numeric missing values (age, price, distance, time, days, total) with median.

Filled categorical missing values (gender, company, flightType, from, to, agency, place, name) with mode or 'Unknown'.

Insight:

Users dataset: Very few missing ages and genders, so imputing median/mode retains most data.

Flights and hotels datasets had missing prices and distances; filling them prevents errors in regression models.

Ensures that all columns are complete for EDA and machine learning.

3. Date Conversion

Manipulation: Converted date columns in flights and hotels datasets to datetime type.

Insight:

Allows extraction of month, weekday, or year, enabling analysis of seasonal travel trends.

Helps answer questions like: “Which month has highest bookings?” or “Weekend vs weekday travel patterns”.

4. Resetting Index

Manipulation: After removing duplicates, reset the DataFrame index.

Insight: Maintains clean, sequential indexing, which is important for merging datasets and iterating through rows.

5. Dataset Ready for Analysis

After wrangling:

No missing values remain.

Numeric and categorical types are consistent.

Linking columns (userCode and travelCode) are intact.

Insight: The datasets are now ready for Exploratory Data Analysis (EDA), regression modeling (flight price prediction), classification (gender prediction), and recommendation system building.

Reasoning:

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀):
There is no significant difference in the average flight prices booked by male and female users.

H₀: μₘₐₗₑ = μ_female

Alternative Hypothesis (H₁):
There is a significant difference in the average flight prices booked by male and female users.

H₁: μₘₐₗₑ ≠ μ_female

#### 2. Perform an appropriate statistical test.

Since we are comparing the means of two independent groups (male vs female) and the variable price is numerical, we perform an Independent Two-Sample t-test (Welch’s t-test).

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

# Merge datasets
merged_df = flights_df.merge(users_df, left_on='userCode', right_on='code')

# Separate price values by gender
male_prices = merged_df[merged_df['gender'] == 'Male']['price']
female_prices = merged_df[merged_df['gender'] == 'Female']['price']

# Welch's t-test
t_stat, p_value = ttest_ind(male_prices, female_prices, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

Independent Two-Sample t-test (Welch’s t-test).

##### Why did you choose the specific statistical test?

- We are comparing the means of two independent groups (male and female).

- The variable flight price is numerical and continuous.

- Variances between groups may be different, therefore Welch’s t-test is more appropriate.

````python
# -------------------------------
# ML Model - 1 Implementation
# Regression Model: Predict Flight Price
# -------------------------------

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, GridSearchCV

# Features and target
X = flights_df[['distance_scaled', 'price_per_km_scaled']]
y = flights_df['price_log']  # log-transformed price

# -------------------------------
# 1. Fit the Algorithm
# -------------------------------
model = LinearRegression()
model.fit(X, y)

# Predict on the model
y_pred = model.predict(X)
````

````python
# Visualizing evaluation Metric Score chart
# -------------------------------
# 2. Evaluate Model Performance
# -------------------------------
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("RMSE:", rmse)
print("R2 Score:", r2)

# Visualizing Evaluation Metric Score chart
import matplotlib.pyplot as plt

metrics = {'RMSE': rmse, 'R2 Score': r2}
plt.bar(metrics.keys(), metrics.values(), color=['skyblue', 'orange'])
plt.title('Regression Model Evaluation Metrics')
plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# -------------------------------
# 3. Cross Validation & Hyperparameter Tuning
# -------------------------------
# Using GridSearchCV as an example (works for regression models with hyperparameters)
from sklearn.linear_model import Ridge

# Hyperparameter grid
param_grid = {'alpha': [0.01, 0.1, 1, 10, 50]}
ridge = Ridge()
grid_search = GridSearchCV(estimator=ridge, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X, y)

# Best model
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X)

# Evaluation after tuning
rmse_best = np.sqrt(mean_squared_error(y, y_pred_best))
r2_best = r2_score(y, y_pred_best)

print("Best alpha:", grid_search.best_params_)
print("RMSE after tuning:", rmse_best)
print("R2 Score after tuning:", r2_best)

# Visualizing Updated Evaluation Metric Score chart
metrics_best = {'RMSE': rmse_best, 'R2 Score': r2_best}
plt.bar(metrics_best.keys(), metrics_best.values(), color=['green', 'red'])
plt.title('Tuned Regression Model Evaluation Metrics')
plt.show()
````
